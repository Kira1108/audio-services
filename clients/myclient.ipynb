```python
import pandas as pd
import json
import copy

with open('result.json', 'r') as file:
    data = json.load(file)
```

```python
results = []
for rec in data:
    if rec.get("action","") == "receive":
        temp = copy.deepcopy(rec)
        temp.update(
            {
                "start_time": rec['data'].get("start_time", ""),
                "end_time": rec['data'].get("end_time", ""),
                "is_partial": rec['data'].get("is_partial", ""),
            }
        )
        results.append(temp)
```

```python
df = pd.DataFrame(results)

df[df.is_partial == False].to_excel("result.xlsx", index=False)
```
